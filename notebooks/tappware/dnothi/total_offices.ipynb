## Caution: All manually created users are not counted (Action will be taken in future)

````python
# SELECT count(id) FROM offices WHERE date(created) <= '2020-09-30' AND active_status =1;

# Extract mandatory attributes for offices graph analysis
print("Loading dataframe ...\n")
offices_df = load_offices_df()
print(f"Initial dataframe shape {offices_df.shape}\n")
# making dataframe short list
offices_df = offices_df[['id', 'active_status', 'created']] 
print("Filtering with active_status = 1")
offices_df = offices_df[offices_df.active_status==1] 
print(f"dataframe shape: {offices_df.shape}\n")
# remove null values
print("Filtering with created != null values")
offices_df = offices_df.loc[offices_df.created.notnull()]
print(f"dataframe shape: {offices_df.shape}\n")
# add new column: cretead_new as datetime field from created column
print("Converting created field to datetime field and substitue not convertable field with null value\n")
offices_df['created'] = pd.to_datetime(offices_df['created'], errors='coerce')
# errors{‘ignore’, ‘raise’, ‘coerce’}, default ‘raise’
    # If ‘raise’, then invalid parsing will raise an exception.
    # If ‘coerce’, then invalid parsing will be set as NaT.
    # If ‘ignore’, then invalid parsing will return the input.
# again remove null values based on created field
print("Filtering with created != null values")
offices_df = offices_df.loc[offices_df.created.notnull()]
print(f"dataframe shape: {offices_df.shape}\n")
````

````python
# Extract years and months from created column
created_datetime_index = pd.DatetimeIndex(offices_df['created'])
years = created_datetime_index.year.values.astype(str)
months = created_datetime_index.month.values.astype(str)
days = created_datetime_index.day.values.astype(str)
# type(years)
# type(months)
````

````python
# type(years[0])
offices_df['year'] = years
offices_df['month'] = months
offices_df['day'] = days
offices_df.shape
````

````python
month_map ={
    '1': 'January',
    '01': 'January',
    '2': 'February',
    '02': 'February',
    '3': 'March',
    '03': 'March',
    '4': 'April',
    '04': 'April',
    '5': 'May',
    '05': 'May',
    '6': 'June',
    '06': 'June',
    '7': 'July',
    '07': 'July',
    '8': 'August',
    '08': 'August',
    '9': 'September',
    '09': 'September',
    '10': 'October',
    '11': 'November',
    '12': 'December',
    'unknown': 'unknown',
}
````

````python
# month_sequence_list =['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August','September', 'October','November', 'December', 'unknown' ]

total_offices = []
year_data = []

dataframe_year_by = offices_df.groupby('year')

for year, year_frame in dataframe_year_by:
    year = str(year)
    # year, year_frame.shape
    year_dict = {}
    year_dict['year'] = year
    year_dict['count'] = int(year_frame.shape[0])
    year_dict['month_data'] = []
    
    month_data = []
    

    month_group_by = year_frame.groupby('month')
    for month, month_frame in month_group_by:
        month_dict = {}
        month_dict['month'] = month
        month_dict['count'] = month_frame.shape[0]
        month_dict['day_data'] = []
        
        day_data = []
        day_group_by = month_frame.groupby('day')
        for day, day_frame in day_group_by:
            day_dict = {}
            day_dict['day'] = day
            day_dict['count'] = day_frame.shape[0]
            day_data.append(day_dict)
        month_dict['day_data'] = day_data
        month_data.append(month_dict)
        
    year_dict['month_data'] = month_data
    year_data.append(year_dict)

# print(total_offices)
# print(year_data)
total_offices = year_data
````

````python
import copy
data = total_offices

MONTH_SEQUENCE_LIST =['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August','September', 'October','November', 'December', 'unknown' ]
DAY_SEQUENCE_LIST = []
initialize_day_list = []
for i in range(32):
    lst = [i, 0]
    initialize_day_list.append(lst)


def generate_general_series_and_drilldown_series(data, general_series_name):
    GENERAL_SERIES = [
        {
            'name': general_series_name,
            'colorByPoint': True,
            'data': [],
        }
    ]
    DRILLDOWN_SERIES = []

    for year_dict in data:
        year = year_dict['year']
        count = year_dict['count']
        month_data = year_dict['month_data']
        
        year = str(year)

        temporary_dict_general = {
            'name': year,
            'y': count,
            'drilldown': year,
        }
        GENERAL_SERIES[0]['data'].append(temporary_dict_general)

        temporary_dict_drilldown = {
            'name': year,
            'id': year,
            'data': [],
        }

        for month_dict in month_data:
            month = month_dict['month']
            count = month_dict['count']
            day_data = month_dict['day_data']
            month = str(month)

            dic = {}
            dic['name'] = month_map[month]
            dic['y'] = count
            dic['drilldown'] = year + month

            temporary_dict_drilldown['data'].append(dic)
            temporary_day_dict = {'id': year + month, 'data': copy.deepcopy(initialize_day_list)}

            for day_dict in day_data:
                day = day_dict['day']
                count = day_dict['count']
                day = int(day)
                
                lst = [day, count]
        
                temporary_day_dict['data'].remove([day, 0])
                temporary_day_dict['data'].append(lst)
     
            # sort days
            sorted_days = sorted(temporary_day_dict['data'], key=lambda x:x[0])
            temporary_day_dict['data'] = sorted_days
            
            DRILLDOWN_SERIES.append(temporary_day_dict)

        # sort month
        sorted_month = sorted(temporary_dict_drilldown['data'], key=lambda x:MONTH_SEQUENCE_LIST.index(x['name']))
        temporary_dict_drilldown['data'] = sorted_month

        DRILLDOWN_SERIES.append(temporary_dict_drilldown)

    return GENERAL_SERIES, DRILLDOWN_SERIES
GENERAL_SERIES, DRILLDOWN_SERIES = generate_general_series_and_drilldown_series(data, 'years')

# DRILLDOWN_SERIES
````
