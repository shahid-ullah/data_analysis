```python
import re
# year_month_regex = re.compile(r'(\d\d\d\d)-(\d\d)-(\d\d)')
year_month_day_regex = re.compile(r'(\d\d\d\d)-(\d\d)-(\d\d)')
```

```python
# users_df.info()
years = []
months = []
days = []
indexes = []

users_df = pd.read_csv('./Data/users.csv', names=users_table_columns)
users_df = users_df.loc[users_df.created.notnull()]
times = users_df.created.values
for index, time in enumerate(times, start=0):
    if time == 'unknown':
        year = time
        month = time
        day = time
        years.append(year)
        months.append(month)
        days.append(day)
        indexes.append(index)
    else:
        # print(time)
        mo = year_month_day_regex.search(time)
        year = mo.group(1)
        month = mo.group(2)
        day = mo.group(3)
        years.append(year)
        months.append(month)
        days.append(day)
        indexes.append(index)
        
# set(years)
# set(months)
# set(days)
x = pd.DatetimeIndex(users_df['created'])
year2 = x.year
years = years[:50]
year2 = year2[:50]
for y1, y2 in zip(years, year2):
    if y1 == y2:
        pass
    else:
        'not same'
        print(y1, y2)
```
